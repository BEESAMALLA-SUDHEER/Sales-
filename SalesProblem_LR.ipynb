```python
'''
Predicting Mean
'''

# importing required libraries
import pandas as pd
from sklearn.metrics import mean_squared_error
```

```python
# read the train and test dataset
train_data = pd.read_csv('train2.csv')
test_data = pd.read_csv('test2.csv')

print(train_data.head())

# shape of the dataset
print('\nShape of training data :',train_data.shape)
print('\nShape of testing data :',test_data.shape)
```

```python
# Now, we need to predict the missing target variable in the test data using the mean only
# target variable - Item_Outlet_Sales

train_y = train_data['Item_Outlet_Sales']

# test_y = test_data['Item_Outlet_Sales']
```

```python
print('Mean of Target Variable : ',train_y.mean())

# predict the target on the test dataset
predict_train = [train_y.mean() for i in range(train_y.shape[0])]
 
# Root Mean Squared Error on training dataset
rmse_train = mean_squared_error(train_y,predict_train)**(0.5)
print('\nMSE on train dataset : ', rmse_train)
```

#### Linear regression with more variables

```python
train = pd.read_csv('train2.csv')

test = pd.read_csv('test2.csv')

# imputing missing values

train['Item_Visibility'] = train['Item_Visibility'].replace(0,np.mean(train['Item_Visibility']))

train['Outlet_Establishment_Year'] = 2013 - train['Outlet_Establishment_Year']

train['Outlet_Size'].fillna('Small',inplace=True)
train['Item_Weight'].fillna((train['Item_Weight'].mean()), inplace=True)

# creating dummy variables to convert categorical into numeric values

mylist = list(train.select_dtypes(include=['object']).columns)

dummies = pd.get_dummies(train[mylist], prefix= mylist)

train.drop(mylist, axis=1, inplace = True)

X = pd.concat([train,dummies], axis =1 )
X
```

```python
X.isnull().sum()
```

```python
X = train.drop('Item_Outlet_Sales',1)
x_train, x_val, y_train, y_val = train_test_split(X,train.Item_Outlet_Sales, test_size =0.3)

lreg= LinearRegression()
# training a linear regression model on train
lreg.fit(x_train,y_train)

# predicting on val
pred_val = lreg.predict(x_val)

# calculating mse
mse = np.mean((pred_val - y_val)**2)
```

```python
mse
```

```python
lreg.score(x_val,y_val)
```
